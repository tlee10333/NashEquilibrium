```python
# Extension
from WS import Network
from networks import alpha, beta
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Erdos Renyi Nowak & May

alphaw = [[], [], [], [], [], [], [], [], [], []]
betaw = [[], [], [], [], [], [], [], [], [], []]

timestep = 500
temptation = 1.1
for s in range(10):
    fig, ax = plt.subplots(1, 2)

    w = Network(1024, 6, 0.1, graph_type="er", t=temptation, r=1, p=0, s=0)

    plt.subplot(1, 2, 1)
    plt.xlabel("Before")
    plt.title("Nowak & May ER Graph with T = " + str(temptation))
    w.subplot_draw()

    for i in range(timestep):
        w.round()
        alphaw[s].append(alpha(w))
        betaw[s].append(beta(w))
        w.update()

    plt.subplot(1, 2, 2)
    plt.xlabel("After")
    w.subplot_draw()
    plt.show()

    print("NEXT")
    temptation += 0.1
```

```python
#ER Graphing NM

label = 1.1
for s in range(10):
    plt.plot([i for i in range(timestep)], alphaw[s], label=str(label)[0:3])
    label += 0.1

plt.xscale("log")
plt.title("Parameter Sweep of T vs. Alpha for Nowak-May ER Graph")
plt.xlabel("Timesteps")
plt.ylabel("Alpha")
plt.legend()
plt.show()
label = 1.1

for s in range(10):
    plt.plot([label for i in range(timestep)],
             betaw[s], "o", label=str(label)[0:3])
    label += 0.1


plt.legend()
plt.title("Beta Distribution for T Values")
plt.xlabel("T")
plt.ylabel("Beta")
plt.show()
```
