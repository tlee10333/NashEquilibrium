```python
from WS import Network
from networks import alpha, beta
import networkx as nx
import matplotlib.pyplot as plt
```

```python
alphaw = []
betaw = []

alphaz = []
betaz = []

alphax = []
betax = []

alphay = []
betay = []

temptation = 1.2
w = Network(
    1024, 6, 0.1, graph_type="ws", t=temptation, r=1, p=0, s=0
)  # Nowak & May WS
z = Network(
    1024, 6, 0.1, graph_type="ba", t=temptation, r=1, p=0, s=0
)  # Nowak & May BA
x = Network(
    1024, 6, 0.1, graph_type="ws", t=temptation, r=1, p=0, s=0
)  # Santos Pacheco WS
y = Network(
    1024, 6, 0.1, graph_type="ba", t=temptation, r=1, p=0, s=0
)  # Santos Pacheco BA


timestep = 10000
for i in range(timestep):
    w.round()
    z.round()
    x.round()
    y.round()
    alphaw.append(alpha(w))
    betaw.append(beta(w))

    alphaz.append(alpha(z))
    betaz.append(beta(z))

    alphax.append(alpha(x))
    betax.append(beta(x))

    alphay.append(alpha(y))
    betay.append(beta(y))

    w.update()
    z.update()
    x.update(1)
    y.update(1)
```

```python
plt.plot([i for i in range(timestep)], alphaw, label="Nowak & May WS", color="cyan")
plt.plot([i for i in range(timestep)], alphaz, label="Nowak & May BA", color="black")
plt.plot(
    [i for i in range(timestep)], alphax, label="Santos & Pacheco WS", color="green"
)
plt.plot([i for i in range(timestep)], alphay, label="Santos & Pacheco BA", color="red")


plt.xscale("log")
plt.title("Comparison Between All 4 Graphs when T=1.2")
plt.xlabel("Timesteps")
plt.ylabel("Alpha")
plt.legend()
plt.show()
```

```python
plt.plot([i for i in range(timestep)], betaw,
         label="Nowak & May WS", color="cyan")
plt.plot([i for i in range(timestep)], betaz,
         label="Nowak & May BA", color="black")
plt.plot(
    [i for i in range(timestep)], betax, label="Santos & Pacheco WS", color="green"
)
plt.plot([i for i in range(timestep)], betay,
         label="Santos & Pacheco BA", color="red")


plt.xscale("log")
plt.title("Comparison Between All 4 Graphs when T=1.2")
plt.xlabel("Timesteps")
plt.ylabel("Beta")
plt.legend()
plt.show()
```
