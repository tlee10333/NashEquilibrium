```python
from WS import Network

import networkx as nx
```

```python
x = Network(1024, 6, 0.1, graph_type="ws")

y = Network(1024, 6, graph_type="ba")
```

```python
x.draw() #WS Graph
print("WS Graph")
y.draw() #BA Graph  
print("BA Graph")
#Red = defector, Blue = cooperator
```

```python
#Nowak & May

for i in range(11000):
    x.round()
    y.round()
    x.update()
    y.update()
```

```python
x.draw() #WS Graph
print("WS Graph")
y.draw() #BA Graph  
print("BA Graph")
```

```python
a = Network(1024, 6, 0.1, graph_type="ws")
b = Network(1024, 6, graph_type="ba")

#Santos-Pacheco

for i in range(11000):
    a.round()
    b.round()
    a.update(1)
    b.update(1)
a.draw() #WS Graph
print("WS Graph")
b.draw() #BA Graph  
print("BA Graph")
```
